```python
# Camera Density Visualization
def plot_camera_density(gdf, level):
    plt.figure(figsize=(10, 6))
    plt.barh(gdf['Code_Name'], gdf['camera_density'], color='skyblue')
    plt.xlabel("Cameras per km²")
    plt.ylabel(f"{level} Districts")
    plt.title(f"Camera Density in {level} Districts")
    plt.tight_layout()
    plt.savefig(f"{output_path}/{level}_camera_density.png")
    plt.show()

# Population vs Camera Counts Scatter Plot
def plot_population_vs_cameras(gdf, level):
    plt.figure(figsize=(8, 6))
    plt.scatter(gdf['2021'], gdf['camera_count'], alpha=0.7, color='coral')
    plt.xlabel("Population")
    plt.ylabel("Number of Cameras")
    plt.title(f"Population vs Camera Counts in {level} Districts")

    # # Set shorter intervals for x and y axes
    # x_ticks = np.arange(0, gdf['2021'].max() + 10000, 10000)  # Adjust interval as needed
    # y_ticks = np.arange(0, gdf['camera_count'].max() + 50, 50)  # Adjust interval as needed
    # plt.xticks(x_ticks, rotation=45)
    # plt.yticks(y_ticks)
    
    plt.tight_layout()
    plt.savefig(f"{output_path}/{level}_population_vs_cameras.png")
    plt.show()

# Population vs Camera Counts Heatmap
def plot_population_vs_cameras_heatmap(gdf, level):
    heatmap_data = gdf[['Code_Name', '2021', 'camera_count']].copy()
    heatmap_data.rename(columns={'2021': 'Population', 'camera_count': 'Camera Count'}, inplace=True)
    heatmap_data.set_index('Code_Name', inplace=True)
    
    # Normalize data for better visualization
    heatmap_data_normalized = heatmap_data.div(heatmap_data.max(axis=0))

    # Plot heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(heatmap_data_normalized, cmap="YlOrRd", annot=True, fmt=".2f", linewidths=0.5, cbar_kws={'label': 'Normalized Values'})
    plt.title(f"Heatmap of Camera Counts vs Population in {level} Districts")
    plt.xlabel("Metrics")
    plt.ylabel(f"{level} Districts")
    plt.tight_layout()
    plt.savefig(f"{output_path}/{level}_camera_counts_vs_population_heatmap.png")
    plt.show()

# Population Density vs Camera Density
def plot_population_vs_camera_density(gdf, level):
    plt.figure(figsize=(8, 6))
    plt.scatter(gdf['population_density'], gdf['camera_density'], alpha=0.7, color='green')
    plt.xlabel("Population Density (people per km²)")
    plt.ylabel("Camera Density (cameras per km²)")
    plt.title(f"Population Density vs Camera Density in {level} Districts")
    plt.tight_layout()
    plt.savefig(f"{output_path}/{level}_population_vs_camera_density.png")
    plt.show()

# Population Density vs Camera Density Heatmap
def plot_population_vs_camera_density_heatmap(gdf, level):
    heatmap_data = gdf[['Code_Name', 'population_density', 'camera_density']].copy()
    heatmap_data.rename(columns={'population_density': 'Population Density', 'camera_density': 'Camera Density'}, inplace=True)
    heatmap_data.set_index('Code_Name', inplace=True)
    
    # Normalize data for better visualization
    heatmap_data_normalized = heatmap_data.div(heatmap_data.max(axis=0))

    # Plot heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(heatmap_data_normalized, cmap="YlGnBu", annot=True, fmt=".2f", linewidths=0.5, cbar_kws={'label': 'Normalized Values'})
    plt.title(f"Heatmap of Camera Density vs Population Density in {level} Districts")
    plt.xlabel("Metrics")
    plt.ylabel(f"{level} Districts")
    plt.tight_layout()
    plt.savefig(f"{output_path}/{level}_camera_density_vs_population_density_heatmap.png")
    plt.show()

def plot_population_vs_camera_density_heatmap(gdf, level):
    heatmap_data = gdf[['Code_Name', 'camera_count', 'camera_density', 'population_density']]
    # Set 'Code_Name' as index to use districts as rows
    heatmap_data.set_index('Code_Name', inplace=True)
    # Normalize the data for better comparison across metrics (optional)
    heatmap_data_normalized = heatmap_data.div(heatmap_data.max(axis=0))
    # Create the heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(heatmap_data_normalized, cmap="YlGnBu", annot=True, fmt=".2f", linewidths=0.5, cbar_kws={'label': 'Normalized Values'})
    # Set titles and labels
    plt.title("Heatmap of Population, Camera Counts, and Density by District")
    plt.xlabel("Metrics")
    plt.ylabel("Districts")
    # Save and show the plot
    plt.tight_layout()
    plt.savefig(f"{output_path}/{level}_metrics_heatmap.png")
    plt.show()

# Bar Plot for Camera Count vs Population
def plot_population_vs_cameras_bar(gdf, level):
    # Sort the data by population for better visualization
    gdf = gdf.sort_values(by='2021').reset_index(drop=True)

    plt.figure(figsize=(10, 6))

    # Plot bars
    bar_width = 0.4
    x = np.arange(len(gdf))

    # Population bars
    plt.bar(x - bar_width/2, gdf['2021'], bar_width, label='Population', color='skyblue')

    # Camera count bars
    plt.bar(x + bar_width/2, gdf['camera_count'], bar_width, label='Camera Count', color='coral')

    # Add labels and title
    plt.xlabel(f"{level} Districts (Index Only)", fontsize=12)
    plt.ylabel("Counts", fontsize=12)
    plt.title(f"Camera Count vs Population in {level} Districts", fontsize=14)

    # Use indices for x-axis ticks and remove district info
    plt.xticks(x, x, fontsize=10)

    # Add a legend
    plt.legend()

    # Save the plot
    plt.tight_layout()
    plt.savefig(f"{output_path}/{level}_camera_count_vs_population_barplot_no_labels.png")
    plt.show()
    
# Average Camera Distance Analysis
def calculate_average_camera_distance(camera_data):
    unique_locations = camera_data.drop_duplicates(subset=['latitude', 'longitude']).reset_index(drop=True)
    unique_distances = [
        geodesic((unique_locations.loc[i, 'latitude'], unique_locations.loc[i, 'longitude']),
                 (unique_locations.loc[i + 1, 'latitude'], unique_locations.loc[i + 1, 'longitude'])).meters
        for i in range(len(unique_locations) - 1)
    ]
    return sum(unique_distances) / len(unique_distances) / 1000 if unique_distances else 0
    
# Adjust duplicate coordinates
def adjust_duplicate_coordinates(data, lat_col, lon_col, epsilon=0.00001):
    seen_coordinates = set()
    for idx, row in data.iterrows():
        lat, lon = row[lat_col], row[lon_col]
        while (lat, lon) in seen_coordinates:
            lat += epsilon
            lon += epsilon
        data.at[idx, lat_col] = lat
        data.at[idx, lon_col] = lon
        seen_coordinates.add((lat, lon))

        # Merge population data
def merge_population_data(gdf, gdf_column, population_data, pop_data_column):
    return gdf.merge(population_data, on=gdf_column, how='left')

# Helper function for spatial join to count cameras in each district
def calculate_camera_counts(district_gdf, id_column):
    districts_with_camera_counts = gpd.sjoin(district_gdf, camera_gdf, how="left", predicate="contains")
    districts_with_camera_counts = districts_with_camera_counts[districts_with_camera_counts['camera_present'] == True]
    camera_counts = districts_with_camera_counts.groupby(id_column).size().reset_index(name='camera_count')
    district_gdf = district_gdf.merge(camera_counts, on=id_column, how='left')
    district_gdf['camera_count'] = district_gdf['camera_count'].fillna(0).astype(int)
    return district_gdf

def remove_leading_zeros(data, column):
    """
    Remove leading zeros from the numeric part of the Code_Name column.
    """
    data[column] = data[column].astype(str)  # Ensure all values are strings
    # Use regex to remove leading zeros from numeric parts
    data[column] = data[column].str.replace(r'^0+', '', regex=True)
    return data
```

```python
camera_data = camera_data[camera_data['camera_present'] == True].reset_index(drop=True)
camera_data['latitude'] = camera_data['latitude'].astype(float)
camera_data['longitude'] = camera_data['longitude'].astype(float)

adjust_duplicate_coordinates(camera_data, 'latitude', 'longitude')

# Convert camera data to a GeoDataFrame
camera_gdf = gpd.GeoDataFrame(
    camera_data,
    geometry=gpd.points_from_xy(camera_data.longitude, camera_data.latitude),
    crs="EPSG:4326"
)
```

```python
major_districts_gdf['major_district'] = major_districts_gdf['tunnus'].astype(str) + ": " + major_districts_gdf['nimi_fi']
major_districts_gdf['Code_Name'] = major_districts_gdf['tunnus'].astype(str) + ": " + major_districts_gdf['nimi_fi'].str.lower()

basic_districts_gdf['basic_district'] = basic_districts_gdf['tunnus'].astype(str) + ": " + basic_districts_gdf['nimi_fi']
basic_districts_gdf['Code_Name'] = basic_districts_gdf['tunnus'].astype(str) + ": " + basic_districts_gdf['nimi_fi'].str.lower()

sub_districts_gdf['major_district'] = sub_districts_gdf['suurpiiri_tunnus'].astype(str) + ": " + sub_districts_gdf['suurpiiri_nimi_fi']
sub_districts_gdf['basic_district'] = sub_districts_gdf['peruspiiri_tunnus'].astype(str) + ": " + sub_districts_gdf['peruspiiri_nimi_fi']
sub_districts_gdf['sub_district'] = sub_districts_gdf['osaalue_tunnus'].astype(str) + ": " + sub_districts_gdf['osaalue_nimi_fi']
sub_districts_gdf['Code_Name'] = sub_districts_gdf['osaalue_tunnus'].astype(str) + ": " + sub_districts_gdf['osaalue_nimi_fi'].str.lower()

population_data['Code_Name'] = population_data['Code_Name'].str.lower()
```

```python
# Remove leading zeros from 'Code_Name' in both datasets
population_data = remove_leading_zeros(population_data, 'Code_Name')
major_districts_gdf = remove_leading_zeros(major_districts_gdf, 'Code_Name')
basic_districts_gdf = remove_leading_zeros(basic_districts_gdf, 'Code_Name')
sub_districts_gdf = remove_leading_zeros(sub_districts_gdf, 'Code_Name')
```

```python
major_districts_gdf = merge_population_data(major_districts_gdf, 'Code_Name', population_data, 'Code_Name')
basic_districts_gdf = merge_population_data(basic_districts_gdf, 'Code_Name', population_data, 'Code_Name')
sub_districts_gdf = merge_population_data(sub_districts_gdf, 'Code_Name', population_data, 'Code_Name')
```

```python
# Calculate land area for each district (assuming CRS is set to a metric system)
major_districts_gdf['area_km2'] = (major_districts_gdf.geometry.to_crs({'init': 'EPSG:3857'}).area / 10**6).round(2)  # Area in square kilometers
major_districts_gdf['area_mi2'] = (major_districts_gdf['area_km2'] * 0.386102).round(2)  # Convert km² to mi²

basic_districts_gdf['area_km2'] = (basic_districts_gdf.geometry.to_crs({'init': 'EPSG:3857'}).area / 10**6).round(2)  # Area in square kilometers
basic_districts_gdf['area_mi2'] = (basic_districts_gdf['area_km2'] * 0.386102).round(2)  # Convert km² to mi²

sub_districts_gdf['area_km2'] = (sub_districts_gdf.geometry.to_crs({'init': 'EPSG:3857'}).area / 10**6).round(2)  # Area in square kilometers
sub_districts_gdf['area_mi2'] = (sub_districts_gdf['area_km2'] * 0.386102).round(2)  # Convert km² to mi²

for gdf in [major_districts_gdf, basic_districts_gdf, sub_districts_gdf]:
    if gdf.crs != camera_gdf.crs:
        gdf.to_crs(camera_gdf.crs, inplace=True)
```

```python
major_districts_gdf = calculate_camera_counts(major_districts_gdf, 'id')

basic_districts_gdf = calculate_camera_counts(basic_districts_gdf, 'id')
basic_districts_gdf = basic_districts_gdf.copy().merge(
    sub_districts_gdf[['major_district', 'peruspiiri_tunnus']],
    left_on='tunnus', right_on='peruspiiri_tunnus', how='left',
    suffixes=('', '')  # Use suffix to distinguish columns if needed
)

sub_districts_gdf = calculate_camera_counts(sub_districts_gdf, 'osaalue_tunnus')
# Dissolve geometries at the sub-district level to get one boundary per sub-district
sub_districts_gdf_dissolved = sub_districts_gdf.dissolve(by='osaalue_tunnus', as_index=False)
sub_districts_gdf_dissolved['camera_count'] = sub_districts_gdf_dissolved.groupby('osaalue_tunnus')['camera_count'].transform('sum')
```

```python
# Generate district-wise summary table
major_district_summary = major_districts_gdf[['major_district', 'area_km2', 'area_mi2', '2021', 'camera_count']]
major_district_summary.columns = ['Major-District', 'Land Area (km²)', 'Land Area (mi²)', 'Population (2021)', 'Number of Cameras']
print(major_district_summary.to_latex(index=False, float_format="%.2f"))
```

```python
basic_district_summary = basic_districts_gdf[['major_district', 'basic_district', 'area_km2', 'area_mi2', '2021', 'camera_count']]
basic_district_summary.columns = ['Major District', 'Basic District', 'Land Area (km²)', 'Land Area (mi²)', 'Population (2021)', 'Number of Cameras']
print(basic_district_summary.to_latex(index=False, float_format="%.2f"))
```

```python
sub_district_summary = sub_districts_gdf_dissolved[['major_district', 'basic_district', 'sub_district', 'area_km2', 'area_mi2', '2021', 'camera_count']]
sub_district_summary.columns = ['Major-District', 'Basic District', 'Sub-District', 'Land Area (km²)', 'Land Area (mi²)', 'Population (2021)', 'Number of Cameras']
print(sub_district_summary.to_latex(index=False, float_format="%.2f"))
```

```python
# Calculate land area and camera counts
for gdf, level in zip([major_districts_gdf, basic_districts_gdf, sub_districts_gdf], ["Major", "Basic", "Sub"]):
    # Convert '2021' column to numeric, handling any non-numeric values gracefully
    gdf['2021'] = gdf['2021'].str.replace(',', '')  # Remove commas if present
    gdf['2021'] = pd.to_numeric(gdf['2021'], errors='coerce')  # Convert to numeric, setting invalid parsing to NaN
    
    # Calculate land area and camera-related metrics
    gdf['area_km2'] = gdf.geometry.to_crs({'init': 'EPSG:3857'}).area / 10**6
    gdf = gpd.sjoin(gdf, camera_gdf, how="left", predicate="contains")
    gdf['camera_count'] = gdf.groupby('id')['geometry'].transform('count')
    gdf['camera_density'] = gdf['camera_count'] / gdf['area_km2']
    
    # Calculate population density
    gdf['population_density'] = gdf['2021'] / gdf['area_km2']
    
    # Print a few rows to verify
    print(f"--- {level} Districts ---")
    print(gdf[['id', 'area_km2', 'camera_count', 'camera_density', 'population_density']].head())

    plot_camera_density(gdf, level)
    plot_population_vs_cameras(gdf, level)
    # plot_population_vs_cameras_bar(gdf, level)
    # plot_population_vs_camera_density(gdf, level)
    # plot_population_vs_camera_density_heatmap(gdf, level)
    # plot_population_vs_cameras_heatmap(gdf, level)
    # plot_population_vs_camera_density_heatmap(gdf, level)
    # break
```

```python
average_unique = calculate_average_camera_distance(camera_data)
average_all = calculate_average_camera_distance(camera_data.sort_values(by=['latitude', 'longitude']))

print(average_unique)
print(average_all)

# Plot Average Camera Distances
plt.figure(figsize=(5, 5))
plt.bar(['Unique Locations', 'All Locations'], [average_unique, average_all], color=['blue', 'orange'])
plt.ylabel("Average Distance (km)")
plt.title("Average Distance Between Cameras")
plt.tight_layout()
plt.savefig(f"{output_path}/average_camera_distances.png")
plt.show()
```
