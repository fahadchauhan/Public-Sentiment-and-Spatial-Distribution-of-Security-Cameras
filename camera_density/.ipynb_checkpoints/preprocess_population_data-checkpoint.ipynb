```python
df = pd.read_csv(population_csv_path)
df
```

```python
df.rename(columns={'Unnamed: 0': 'Code_Name'}, inplace=True)
df
```

```python
df = df.dropna(how='all')
df
```

```python
df = df[~df['Code_Name'].astype(str).str.contains('%')]
df.head(10)
df
```

```python
# Display sorted unique values for each column with special characters at the start
for column in df.columns:
    unique_values = df[column].astype(str).unique()  # Ensure all values are treated as strings
    sorted_unique_values = sorted(unique_values, key=lambda x: (not any(c in x for c in '!@#$%^&*()_+-={}[]|\:;\'"<>,.?/~`'), x))
    
    print(f"Sorted unique values in column '{column}':")
    print(sorted_unique_values)
    print("\n")  # Add a newline for better readability
```

```python
replacements = {
    "Vironniemen": "VIRONNIEMI",
    "Ullanlinnan": "ULLANLINNA",
    "Kampinmalmin": "KAMPINMALMI",
    "Taka-Töölön": "Taka-Töölö",
    "Lauttasaaren": "Lauttasaari",
    "Reijolan": "Reijola",
    "Munkkiniemen": "MUNKKINIEMI",
    "Haagan": "Haaga",
    "Pitäjänmäen": "PITÄJÄNMÄKI",
    "Kallion": "Kallio",
    "Alppiharjun": "Alppiharju",
    "Vallilan": "Vallila",
    "Pasilan": "Pasila",
    "Vanhakaupungin": "Vanhakaupunki",
    "Maunulan": "Maunula",
    "Länsi-Pakilan": "Länsi-Pakila",
    "Tuomarinkylän": "Tuomarinkylä",
    "Oulunkylän": "Oulunkylä",
    "Itä-Pakilan": "Itä-Pakila",
    "Latokartanon": "Latokartano",
    "Pukinmäen": "Pukinmäki",
    "Malmin": "Malmi",
    "Suutarilan": "Suutarila",
    "Puistolan": "Puistola",
    "Jakomäen": "Jakomäki",
    "Kulosaaren": "Kulosaari",
    "Herttoniemen": "Herttoniemi",
    "Laajasalon": "Laajasalo",
    "Vartiokylän": "Vartiokylä",
    "Myllypuron": "Myllypuro",
    "Mellunkylän": "Mellunkylä",
    "Vuosaaren": "Vuosaari",
    "Östersundomin": "Östersundom",
    "Kaarelan": "Kaarela",
    "Herttonimenranta": "Herttoniemenranta"
}
```

```python
# Function to parse and classify rows based on both code and name
def classify_row(row):
    code_name = row['Code_Name']

    if '(' in code_name and ')' in code_name:
        # Replace the second-to-last '-' by splitting and joining
        parts = code_name.rsplit('-', 2)
        main_part = '-'.join(parts[:-2]).strip()  # Join up to the second-to-last '-'
    else:
        # Split on the last occurrence of '-' if no brackets are present
        main_part = code_name.rsplit('-', 1)[0].strip()
    
    # Split on spaces to separate the code and name
    parts = [part for part in main_part.split() if part.lower() not in ["peruspiiri", "suurpiiri"]]

    
    # Check if we have at least one part for code
    if parts:
        code = int(re.sub(r'\D', '', parts[0])) if re.sub(r'\D', '', parts[0]) else None
        name = ' '.join(parts[1:])
    else:
        code = None
        name = ''
    
    # Apply replacements
    name = replacements.get(name, name)
#     print(f'code: {code}')
#     print(f'code: {type(code)}')
    # Check if code and name match a Major District, District, or Sub-District
    if code in major_districts.index and major_districts.loc[code, 'hel:suurpiiri_nimi_fi'].lower() == name.lower():
        category = 'Major District'
    elif code in districts.index and districts.loc[code, 'hel:peruspiiri_nimi_fi'].lower() == name.lower():
        category = 'District'
    elif code in sub_districts.index and sub_districts.loc[code, 'hel:osaalue_nimi_fi'].lower() == name.lower():
        category = 'Sub-District'
    else:
        category = 'Unknown'
        print(f'code: {code}')
        print(f"name.lower: {name.lower()}")
        print(f'parts: {parts}')
    
    return pd.Series([code, name, category])

# Apply classification function to each row
df[['Code', 'Name', 'Category']] = df.apply(classify_row, axis=1)

# Display the updated DataFrame
df
```

```python
# Display rows where 'Category' is 'Unknown'
unknown_category_rows = df[df['Category'] == 'Unknown']
print(unknown_category_rows[['Code_Name', 'Code', 'Name', 'Category']])
```

```python
# Update 'Code_Name' column with 'Code:Name' format
df['Code_Name'] = df['Code'].astype(str) + ": " + df['Name']
df
```
